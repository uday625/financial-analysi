```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,10)
y = x**2
```

```python
# Functional 
plt.plot(x,y,'r-')
plt.xlabel("number")
plt.ylabel('Square of the number')
plt.title('Polynomial')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

# looping through axes
# for current in axes:
#     current.plot(x,y)

axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'b')
axes[1].set_title('Second Plot')

plt.tight_layout()
    
#axes.plot(x,y)
```

```python
# Figure Size and DPI
fig = plt.figure(figsize=(5,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout
```

```python
fig.savefig('picture.png',dpi=200)
```

```python
## add label and title
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Figure Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
```

```python
#lagends
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'g', label='x square')
ax.plot(x,x**3,'r', label ='x cube')
ax.plot(x,x, label='x=x')

ax.legend(loc=0)

ax.set_title('Figure Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
```
